```python
기업들=['팀프레시','카카오','핀인사이트']
for 기업 in 기업들:
    URL = 'https://www.jobkorea.co.kr/Search/?stext='+기업+'&tabType=corp&Page_No=1'
    import requests
    응답 = requests.get(URL,headers = 
    {'User-Agent':'Mozilla/5.0'})
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(응답.text, 'html.parser')
    company_tag = soup.select_one('.corp-info .list-post')
    span_tags = company_tag.select_one('.corp').select('span')
    for span in span_tags:
        print(span.text)
        
    




URL = "https://www.jobkorea.co.kr/Search/?stext=%ED%95%98%EB%82%98%EC%9D%80%ED%96%89&tabType=corp&Page_No=1"
```
